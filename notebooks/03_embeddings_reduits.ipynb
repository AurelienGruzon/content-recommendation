```python
from pathlib import Path
import pickle
import numpy as np
from sklearn.decomposition import PCA

BASE_DIR = Path().resolve().parent  # remonte à content-recommendation
DATA_DIR = BASE_DIR / "data/news-portal-user-interactions-by-globocom"

with open(DATA_DIR / "articles_embeddings.pickle", "rb") as f:
    embeddings = pickle.load(f)

print("shape original:", embeddings.shape)

pca = PCA(n_components=50, random_state=42)
embeddings_reduced = pca.fit_transform(embeddings)

print("shape reduced:", embeddings_reduced.shape)

output_dir = Path(BASE_DIR / "azure_function/data")
output_dir.mkdir(parents=True, exist_ok=True)

np.save(output_dir / "articles_embeddings_reduced.npy", embeddings_reduced)
```
